```python
#Step 1: Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import warnings
warnings.filterwarnings('ignore')

# Set styling for plots
plt.style.use('fivethirtyeight')
sns.set_palette('bright')
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
#Step 2: Loading Dataset
df = pd.read_csv('Power_Failures_India_Augmented1.csv')

print("Dataset shape:", df.shape)
df.head()
df.dtypes
df.isnull().sum()
```

```python
#Step 3: Data Cleaning and Preprocessing
df['Failure Severity'].value_counts(dropna=True)

mostCommonSeverity = df['Failure Severity'].mode()[0]
df['Failure Severity'].fillna(mostCommonSeverity, inplace=True)

df.isnull().sum()

#checking duplicates
duplicate_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")
if duplicate_count > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. New shape: {df.shape}")

df.describe()
```

```python
print("\nStep 3: Exploratory Data Analysis (EDA)")

# 1. Distribution of target variable - Failure Occurred
plt.figure(figsize=(10, 6))
sns.countplot(x='Failure Occurred', data=df, palette='viridis')
plt.title('Distribution of Power Grid Failures', fontsize=15)
plt.xlabel('Failure Occurred (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
print(f"Failure rate: {df['Failure Occurred'].mean():.2%}")
```

```python
# 2. Relationship between Grid Load and Failures
plt.figure(figsize=(12, 6))
sns.boxplot(x='Failure Occurred', y='Grid Load (MW)', data=df, palette='Set2')
plt.title('Grid Load Distribution by Failure Status', fontsize=15)
plt.xlabel('Failure Occurred', fontsize=12)
plt.ylabel('Grid Load (MW)', fontsize=12)
plt.show()
```

```python
# 3. Voltage Level and Failures
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Voltage Level (kV)', hue='Failure Occurred', data=df, palette='coolwarm')
plt.title('Failures by Voltage Level', fontsize=15)
plt.xlabel('Voltage Level (kV)', fontsize=12)
plt.ylabel('Count', fontsize=12)
```

```python
# 4. Weather conditions and failures
plt.figure(figsize=(14, 7))
sns.countplot(y='Weather', hue='Failure Occurred', data=df, palette='viridis')
plt.title('Failures by Weather Condition', fontsize=15)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Weather Condition', fontsize=12)
plt.legend(title='Failure Occurred', loc='upper right')
plt.show()
```

```python
# 5. Failure causes distribution
plt.figure(figsize=(14, 8))
sns.countplot(y='Cause of Failure', data=df, palette='muted', order=df['Cause of Failure'].value_counts().index)
plt.title('Distribution of Failure Causes', fontsize=15)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Cause of Failure', fontsize=12)
plt.show()
```

```python
# 6. Location-based failure analysis
plt.figure(figsize=(14, 8))
loc_failure = df.groupby('Location')['Failure Occurred'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='Location', y='Failure Occurred', data=loc_failure, palette='rocket')
plt.title('Failure Rate by Location', fontsize=15)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Failure Rate', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Urban vs Rural failure comparison
plt.figure(figsize=(10, 6))
sns.countplot(x='Urban/Rural', hue='Failure Occurred', data=df, palette='Set1')
plt.title('Urban vs Rural Failures', fontsize=15)
plt.xlabel('Area Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# 8. Correlation heatmap for numerical features
plt.figure(figsize=(12, 8))
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation = numerical_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=15)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# 9. Outage Duration Distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Outage Duration (min)', bins=30, kde=True, color='purple')
plt.title('Distribution of Outage Duration', fontsize=15)
plt.xlabel('Outage Duration (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# 10. Failure Severity and Outage Duration
plt.figure(figsize=(12, 6))
sns.boxplot(x='Failure Severity', y='Outage Duration (min)', data=df, palette='rainbow')
plt.title('Outage Duration by Failure Severity', fontsize=15)
plt.xlabel('Failure Severity', fontsize=12)
plt.ylabel('Outage Duration (minutes)', fontsize=12)
plt.show()
```

```python
print("\nStep 4: Statistical Analysis")
# Group statistics by categorical variables
# 1. Mean outage duration by failure type
print("\nMean Outage Duration by Failure Type:")
print(df.groupby('Failure Type')['Outage Duration (min)'].mean().sort_values(ascending=False))

# 2. Failure rate by location
print("\nFailure Rate by Location:")
print(df.groupby('Location')['Failure Occurred'].mean().sort_values(ascending=False))

# 3. Failure rate by weather condition
print("\nFailure Rate by Weather Condition:")
print(df.groupby('Weather')['Failure Occurred'].mean().sort_values(ascending=False))

# 4. Mean grid load during failures vs non-failures
print("\nMean Grid Load (MW):")
print(df.groupby('Failure Occurred')['Grid Load (MW)'].mean())

# 5. Chi-square test for categorical variables
print("\nChi-square test for Weather vs Failure Occurred:")
from scipy.stats import chi2_contingency
contingency = pd.crosstab(df['Weather'], df['Failure Occurred'])
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"Chi2 value: {chi2:.2f}, p-value: {p:.4f}")
```

```python
print("\nStep 5: Categorical Encoding")
# Identify categorical columns
categorical_cols = ['Location', 'Urban/Rural', 'Weather', 'Failure Type', 'Cause of Failure', 'Failure Severity']

# Create a copy of the dataframe for model training
model_df = df.copy()

# Encode categorical variables
print("Encoding categorical variables...")
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    model_df[col] = le.fit_transform(model_df[col])
    label_encoders[col] = le
    print(f"Encoded {col}: {list(zip(le.classes_, range(len(le.classes_))))}")

print("\nDataset after encoding:")
print(model_df.head())
```

```python
print("\nStep 6: Splitting dataset into train and test sets")
# Define features and target
X = model_df.drop('Failure Occurred', axis=1)
y = model_df['Failure Occurred']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
print("\nStep 7: Feature Scaling (Standardization)")
# Apply standard scaling on numerical features
scaler = StandardScaler()
numerical_features = ['Grid Load (MW)', 'Voltage Level (kV)', 'Outage Duration (min)']
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])
```

```python
print("\nStep 8: Training ML Models")
# 1. Logistic Regression
print("\nTraining Logistic Regression model...")
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")
print("\nClassification Report (Logistic Regression):")
print(classification_report(y_test, lr_pred))

# 2. Decision Tree
print("\nTraining Decision Tree model...")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")
print("\nClassification Report (Decision Tree):")
print(classification_report(y_test, dt_pred))

# 3. Random Forest
print("\nTraining Random Forest model...")
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
print("\nClassification Report (Random Forest):")
print(classification_report(y_test, rf_pred))
```

```python
print("\nStep 9: Feature Importance Analysis")
# Plot feature importance for Random Forest
plt.figure(figsize=(12, 8))
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title('Feature Importance from Random Forest', fontsize=15)
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.show()

print("\nFeature importance ranking:")
print(feature_importance)
```

```python
print("\nStep 10: Converting Predictions to Binary and Visualizing Confusion Matrices")
# Visualize confusion matrices
plt.figure(figsize=(16, 5))

# Logistic Regression
plt.subplot(1, 3, 1)
lr_cm = confusion_matrix(y_test, lr_pred)
sns.heatmap(lr_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Logistic Regression\nConfusion Matrix', fontsize=12)
plt.xlabel('Predicted Label', fontsize=10)
plt.ylabel('True Label', fontsize=10)

# Decision Tree
plt.subplot(1, 3, 2)
dt_cm = confusion_matrix(y_test, dt_pred)
sns.heatmap(dt_cm, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title('Decision Tree\nConfusion Matrix', fontsize=12)
plt.xlabel('Predicted Label', fontsize=10)
plt.ylabel('True Label', fontsize=10)

# Random Forest
plt.subplot(1, 3, 3)
rf_cm = confusion_matrix(y_test, rf_pred)
sns.heatmap(rf_cm, annot=True, fmt='d', cmap='Reds', cbar=False)
plt.title('Random Forest\nConfusion Matrix', fontsize=12)
plt.xlabel('Predicted Label', fontsize=10)
plt.ylabel('True Label', fontsize=10)

plt.tight_layout()
plt.show()

# Model comparison
model_accuracy = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest'],
    'Accuracy': [lr_accuracy, dt_accuracy, rf_accuracy]
}).sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Model', y='Accuracy', data=model_accuracy, palette='viridis')
plt.title('Model Accuracy Comparison')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0, 1)
for i, v in enumerate(model_accuracy['Accuracy']):
    plt.text(i, v + 0.01, f"{v:.4f}", ha='center', fontsize=12)
plt.show()

print("\nBest performing model:", model_accuracy.iloc[0]['Model'])
```

```python
# Save the best model for future use
import joblib
best_model = rf_model if rf_accuracy >= max(dt_accuracy, lr_accuracy) else dt_model if dt_accuracy >= lr_accuracy else lr_model
joblib.dump(best_model, 'power_grid_failure_model.pkl')
print("\nBest model saved as 'power_grid_failure_model.pkl'")

print("\nProject Complete!")
```

```python
print("\nStep 10: Overall Model Performance Evaluation")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix

# Collect all model predictions
models = {
    'Logistic Regression': lr_model,
    'Decision Tree': dt_model,
    'Random Forest': rf_model,
}

# Initialize performance metrics dictionary
performance = {
    'Model': [],
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': [],
    'AUC': []
}

# Calculate metrics for each model
for name, model in models.items():
    # Get predictions and probabilities (if available)
    y_pred = model.predict(X_test)
    
    # For probabilities (needed for ROC curve)
    try:
        if hasattr(model, 'predict_proba'):
            y_pred_proba = model.predict_proba(X_test)[:, 1]
        else:
            y_pred_proba = model.decision_function(X_test)  # For SVM
    except:
        y_pred_proba = y_pred  # Fallback
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Calculate AUC if probabilities are available
    try:
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        auc_score = auc(fpr, tpr)
    except:
        auc_score = np.nan
    
    # Append results
    performance['Model'].append(name)
    performance['Accuracy'].append(accuracy)
    performance['Precision'].append(precision)
    performance['Recall'].append(recall)
    performance['F1 Score'].append(f1)
    performance['AUC'].append(auc_score)

# Create DataFrame from results
performance_df = pd.DataFrame(performance)
performance_df = performance_df.sort_values('F1 Score', ascending=False)

# Display table
print("\nModel Performance Comparison:")
print(performance_df.to_string(index=False, float_format=lambda x: f"{x:.4f}" if not np.isnan(x) else "N/A"))

# Identify best model
best_model_name = performance_df.iloc[0]['Model']
print(f"\nBest model based on F1 Score: {best_model_name}")

# Create visualizations
# 1. Bar chart comparison
plt.figure(figsize=(12, 8))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
df_plot = performance_df.melt(id_vars='Model', value_vars=metrics, var_name='Metric', value_name='Score')

sns.barplot(x='Model', y='Score', hue='Metric', data=df_plot, palette='viridis')
plt.title('Model Performance Comparison', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1.05)
plt.legend(title='Metric', loc='lower right')
plt.tight_layout()
plt.show()

# 4. Confusion Matrix for best model
best_model = models[best_model_name]
y_pred_best = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title(f'Confusion Matrix - {best_model_name}', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.tight_layout()
plt.show()

# Print summary
print("\nPerformance Summary:")
print(f"1. Best overall model: {best_model_name}")
print(f"2. Highest accuracy: {performance_df['Accuracy'].max():.4f} ({performance_df.loc[performance_df['Accuracy'].idxmax(), 'Model']})")
print(f"3. Highest precision: {performance_df['Precision'].max():.4f} ({performance_df.loc[performance_df['Precision'].idxmax(), 'Model']})")
print(f"4. Highest recall: {performance_df['Recall'].max():.4f} ({performance_df.loc[performance_df['Recall'].idxmax(), 'Model']})")
print(f"5. Highest F1 score: {performance_df['F1 Score'].max():.4f} ({performance_df.loc[performance_df['F1 Score'].idxmax(), 'Model']})")

# Calculate statistical significance between top two models (if applicable)
from scipy import stats

top_model = models[performance_df.iloc[0]['Model']]
second_model = models[performance_df.iloc[1]['Model']]

top_pred = top_model.predict(X_test)
second_pred = second_model.predict(X_test)
```
